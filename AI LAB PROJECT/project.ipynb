# **Random Forest Regression Model Application:**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
```

```python
data = pd.read_csv(r"C:\Users\zbook\Desktop\Smart study planner project\cleaned_data.csv")
```

```python
x = data.drop("final_grade", axis=1)   
y = data["final_grade"]  
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestRegressor(
    n_estimators=200,   
    max_depth=15,       
    random_state=42
    )
model.fit(train_x, train_y)

# Predict
predictions = model.predict(test_x)
```

```python
rmse = np.sqrt(mean_squared_error(test_y, predictions))
mae = mean_absolute_error(test_y, predictions)
r2 = r2_score(test_y, predictions)

print("RMSE:", rmse)
print("MAE:", mae)
print("R² Score:", r2)
```

```python
import pandas as pd
import numpy as np
import pickle
import joblib

# Load Model & Encoder (agar tumne encoders save kiye hain)
model = pickle.load(open('student_model.pkl', 'rb'))
encoders = joblib.load('encoders.pkl')

# Encoding helper
def encode_cols(le, value):
    if value in le.classes_:
        return le.transform([value])[0]
    else:
        return -1   # unknown value

# User Inputs
print("Enter Student Data:")

user_input = {
    "study_hours_per_week": float(input("Study Hours per Week: ")),
    "sleep_hours_per_day":  float(input("Sleep Hours per Day: ")),
    "attendance_percentage": float(input("Attendance Percentage: ")),
    "assignments_completed": int(input("Assignments Completed: ")),
    "participation_level":   input("Participation Level (Low/Medium): "),
    "internet_access":       input("Internet Access (Yes/No): "),
    "parental_education":    input("Parental Education (High School/Master's/PhD): "),
    "extracurricular":       input("Extracurricular (Yes/No): "),
    "part_time_job":         input("Part-time Job (Yes/No): ")
}

# One-hot encoding style mapping
features = [
    'study_hours_per_week', 'sleep_hours_per_day', 'attendance_percentage',
    'assignments_completed','participation_level_Low', 'participation_level_Medium',
    'internet_access_Yes',
    'parental_education_High School', "parental_education_Master's",
    'parental_education_PhD',
    'extracurricular_Yes', 'part_time_job_Yes'
]

row = {col: 0 for col in features}

# Fill numeric
row['study_hours_per_week'] = user_input['study_hours_per_week']
row['sleep_hours_per_day'] = user_input['sleep_hours_per_day']
row['attendance_percentage'] = user_input['attendance_percentage']
row['assignments_completed'] = user_input['assignments_completed']

# Fill categorical
if user_input['participation_level'] == 'Low':
    row['participation_level_Low'] = 1
elif user_input['participation_level'] == 'Medium':
    row['participation_level_Medium'] = 1

if user_input['internet_access'] == 'Yes':
    row['internet_access_Yes'] = 1

if user_input['parental_education'] == 'High School':
    row['parental_education_High School'] = 1
elif user_input['parental_education'] == "Master's":
    row["parental_education_Master's"] = 1
elif user_input['parental_education'] == 'PhD':
    row['parental_education_PhD'] = 1

if user_input['extracurricular'] == 'Yes':
    row['extracurricular_Yes'] = 1

if user_input['part_time_job'] == 'Yes':
    row['part_time_job_Yes'] = 1


# Convert to DataFrame
x = pd.DataFrame([row])

# Predict
predict = model.predict(x)
print("Predicted Result:", predict[0])
```
